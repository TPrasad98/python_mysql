```python
import csv
with open("car.data",'r') as f :
    car_data=csv.reader(f,delimiter="\n")
    print(car_data)
    query="""insert into cardataset.car(Buying,Maint,Doors,Person,Lug_boot,Safety,Class_Label) values(%s,%s,%s,%s,%s,%s,%s)"""
    
    for i in car_data:
        split=str(i[0]).split(",")
        
        
        cursor.execute(query,split)
        
    connection.commit()
```
